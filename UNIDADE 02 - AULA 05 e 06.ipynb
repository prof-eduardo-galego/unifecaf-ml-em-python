```python
dados = {'Nome': ['Galhad', 'Aranmi', 'Faoguor', 'Taranor'],
         'Força': [3, 4, 7, 14],
         'Habilidade': ['Espada', 'Espada', 'Arco e Flecha', 'Magia']}
  
df_lor = pd.DataFrame(dados)
df_lor
```

```python
df_lor['tamanho_nome'] = df_lor['Nome'].str.len()
df_lor['Força após treinamento'] = df_lor['Força'] * 3
df_lor
```
